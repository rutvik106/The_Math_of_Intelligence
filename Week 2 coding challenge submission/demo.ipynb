```python
from numpy import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(m, b, data):
    totalError = 0.0
    numInstances = data.shape[0]

    for i in range(numInstances):
        happinessScore = data[i, 0]  # Row 'i' column 'HappinessScore'
        freedom = data[i, 1]  # Row 'i' column 'Freedom'

        currentTarget = freedom

        # Predicted rating with our current fitting line
        # y = mx + b
        currentOutput = m * happinessScore + b

        # Compute squared error
        currentSquaredError = (currentTarget - currentOutput) ** 2

        # Add it to the total error
        totalError += currentSquaredError

    sse = totalError / numInstances

    return sse
```

```python
def gradient_descent_step(m, b, data, learningRate):
    N = data.shape[0]
    m_grad = 0
    b_grad = 0

    for i in range(N):
        # Get current pair (x,y)
        x = data[i, 0]
        y = data[i, 1]

        # Partial derivative respect 'm'
        dm = -((2 / N) * x * (y - (m * x + b)))

        # Partial derivative respect 'b'
        db = - ((2 / N) * (y - (m * x + b)))

        # Update gradient
        m_grad = m_grad + dm
        b_grad = b_grad + db

    # Set the new 'better' updated 'm' and 'b'
    m_updated = m - learningRate * m_grad
    b_updated = b - learningRate * b_grad

    return m_updated, b_updated
```

```python
def compute_total_error(m, b, data):  # Computes total mean squared error
    totalError = 0
    for i in range(len(data)):
        x = data[i, 0]
        y = data[i, 1]
        totalError += (y - (
            m * x + b)) ** 2  # Error is calculated as y' = mx + b(Assuming linear regression) so E = (y-y')^2, summed over all points
    # print("Error: ", totalError / float(len(data)))
    return totalError / float(len(data))  # Returning the mean squared error.
```

```python
def total_error(point_pair, data):  # driver function for compute_total_error
    return compute_total_error(point_pair[0], point_pair[1], data)
```

```python
def compute_jacobian(point_pair, data, h=1e-5):  # computes the jacobian of the function total_error
    n = len(point_pair)
    jacobian = zeros(n)  # initialize the jacobian matrix
    for i in range(n):
        x_i = zeros(n)
        x_i[i] += h  # add the limit value, any small value > 0 should do
        jacobian[i] = (total_error((point_pair + x_i), data) - total_error(point_pair,
                                                                           data)) / h  # calculate derivative using first principle method f'(x) = lt(h->0) (f(x+h) - f(x))/h
    return jacobian  # return the jacobian for the pair of points
```

```python
def compute_hessian(point_pair, data,
                    h=1e-5):  # computes the hessian of the function total_error, it is found as the derivative of the jacobian
    n = len(point_pair)
    hessian = zeros((n, n))  # initialize the hessian matrix
    for i in range(n):
        x_i = zeros(n)
        x_i[i] += h  # add the limit value, any small value > 0 should do
        hessian[i] = (compute_jacobian((point_pair + x_i), data) - compute_jacobian(
            point_pair,
            data)) / h  # calculate derivative using first principle method f'(x) = lt(h->0) (f(x+h) - f(x))/h

    return hessian  # return the jacobian for the pair of points
```

```python
def compute_newton(init_points, data, max_iter=100,
                   e=1e-5):  # calculate roots of the equation, i.e. find x if f(x) = 0. In our case we want to find the minima point, so we find f'(x) = 0
    point_pair_arr = zeros((max_iter, len(init_points)))  # initalize m,b values
    point_pair_arr[0] = init_points  # start points
    opt_val = None  # optimal_value to return
    for i in range(max_iter):
        jacobian = compute_jacobian(point_pair_arr[i], data)  # calculate the jacobian at current m,b
        hessian = compute_hessian(point_pair_arr[i], data)  # calculate the hessian at current m,b
        point_pair_arr[i + 1] = point_pair_arr[i] - dot(linalg.pinv(hessian),
                                                        jacobian)  # calulate the new m, new b using newton's equation x(t+1) = x(t) - f(x(t))/f'(x(t)) but we want to find root of f'(x) so we would do x(t+1) = x(t) - f'(x(t))/f''(x(t))
        opt_val = point_pair_arr[i + 1]
        if abs(total_error(point_pair_arr[i + 1], data) - total_error(
                point_pair_arr[i], data)) < e:  # used for early stopping, stops when there is no real improvement.
            print('Optimal m is %.6f and Optimal b is %.6f - Error: %.6f' % (
                point_pair_arr[i + 1, 0], point_pair_arr[i + 1, 1], total_error(point_pair_arr[i + 1], data)))
            break

    return opt_val
```

```python
def predict(inter, slope):
    freedom = int(input("Enter % Of Freedom(0-100): "))

    if (100 > freedom < 0):
        print("Invalid Input")
    else:
        happiness = (slope * (freedom / 100) + inter) * 1000
        print("The predicted happiness is: {0}".format(happiness))
        predict(inter, slope)
```

```python
def run():
    csv_data = pd.read_csv("2017.csv")
    csv_data.head()

    data = csv_data.as_matrix()

    m = 0
    b = 0
    learningRate = 0.0001
    error = compute_error_for_line_given_points(m, b, data)

    print('For the fitting line: y = %sx + %s\nSSE: %.6f' % (m, b, error))

    for i in range(100):
        m, b = gradient_descent_step(m, b, data, learningRate)
        error = compute_error_for_line_given_points(m, b, data)
        # print('At step %d - Line: y = %.6fx + %.6f - Error: %.6f' % (i + 1, m, b, error))

    print("1'st order optimization")
    print('After 100 Iterations Best fit line: y = %.6fx + %.6f - Error: %.6f' % (m, b, error))

    plt.scatter(x=data[:, 0], y=data[:, 1])
    plt.xlabel("Freedom")
    plt.ylabel("Happiness Score")
    # plt.show()

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_title('Freedom vs Happiness Score (1st Order Optimization)')
    ax.scatter(x=data[:, 0], y=data[:, 1], label='Data')
    plt.plot(data[:, 0], m * data[:, 0] + b, color='red', label='Our Fitting Line')
    ax.set_xlabel('Freedom')
    ax.set_ylabel('Happiness Score')
    ax.legend(loc='best')

    init_points = array([0.0, 1.0])
    newton_points = compute_newton(init_points, data, 100)

    print("2'nd order optimization")
    print('After 100 Iterations Best fit line: y = %.6fx + %.6f - Error: %.6f' % (m, b, error))

    b = newton_points[1]
    m = newton_points[0]

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_title('Freedom vs Happiness Score (2nd Order Optimization)')
    ax.scatter(x=data[:, 0], y=data[:, 1], label='Data')
    plt.plot(data[:, 0], newton_points[0] * data[:, 0] + newton_points[1], color='red', label='Our Fitting Line')
    ax.set_xlabel('Freedom')
    ax.set_ylabel('Happiness Score')
    ax.legend(loc='best')

    plt.show()

    predict(b, m)
```

```python
run()
```
